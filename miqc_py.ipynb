```python
from anndata import read_h5ad
import numpy as np
import pandas as pd
import scanpy as sc
from os.path import join
import altair as alt
```

```python
adata = read_h5ad("zeisel-brain.h5ad")
```

```python
from stepmix.stepmix import StepMix
```

```python
Y = adata.obs["subsets_mito_percent"].values
X = np.expand_dims(adata.obs["detected"].values, axis=1)
```

```python
# Categorical StepMix Model with 2 latent classes
model = StepMix(n_components=2, measurement="covariate", assignment="soft")
model.fit(X, Y)
```

```python
model.get_parameters_df()
```

```python
posterior_cols = model.predict_proba(X, Y)
posterior_cols
```

```python
df = pd.DataFrame(posterior_cols)
alt.Chart(df).mark_point().encode(
    x="0:Q",
    y="1:Q"
)
```
